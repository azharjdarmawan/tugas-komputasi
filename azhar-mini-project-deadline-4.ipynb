# Metode Regressi

### Jumlah Titik Api vs. Curah Hujan

```julia
plot_tipi_vs_cujan = scatter(cujan,tipi, color="blue", xaxis="Curah Hujan", markersize=5, yaxis="Jumlah Titik Api")
```

```julia
tipi_vs_cujan = DataFrame(cujan = cujan, tipi = tipi)
```

```julia
reg_tipi_vs_cujan = lm(@formula(tipi ~ cujan), tipi_vs_cujan)
```

```julia
coef(reg_tipi_vs_cujan)
```

```julia
l_fit = predict(reg_tipi_vs_cujan)
plot(cujan,l_fit, color="red", markersize=5)
scatter!(cujan,tipi, color="blue", xaxis="Curah Hujan", markersize=5, yaxis="Jumlah Titik Api")
```
